```python
wap = wap[(wap['message'].astype(str).str.contains('diccionario'))|
          (wap['message'].astype(str).str.contains('autocorrect'))|
          (wap['message'].astype(str).str.contains('*', regex=False))|
          (wap['message'].astype(str).str.contains('auto correct'))]
```

```python
wap
```

```python
counted = pd.DataFrame(columns=['victor','gaby','Loo','val','joe'])

counted['victor'] = wap[wap['sender'] == 'Victor Rumay'].count()
counted['gaby'] = wap[wap['sender'] == 'Gaby Lechin'].count()
counted['loo'] = wap[wap['sender'] == 'Loo Nava'].count()
counted['val'] = wap[wap['sender'] == 'Valerie Barbich'].count()
counted['joe'] = wap[wap['sender'] == 'Jose Luis Espinoza'].count()

```
